```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
pd.options.display.max_columns = 50
```

```python
math = pd.read_csv("data/student-mat.csv")
math.head(3)
```

```python
port = pd.read_csv("data/student-por.csv")
port.head(3)
```

```python
math.rename(columns={'G1':'G1_Mat', 'G2':'G2_Mat', 'G3':'G3_Mat'}, inplace=True)
port.rename(columns={'G1':'G1_Por', 'G2':'G2_Por', 'G3':'G3_Por'}, inplace=True)
```

```python
conc = pd.merge(math, port, on=['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences'])
conc = conc.drop_duplicates()
conc.head(3)
```
